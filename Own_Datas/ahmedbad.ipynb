```python
url='https://www.99acres.com/search/property/buy/mumbai?city=12&preference=S&area_unit=1&budget_min=0&res_com=R&isPreLeased=1'
source=[]
# #driver=webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
driver.get(url)
```

```python
for i in range(0,2244):
    html = driver.page_source
    driver.find_element(by=By.XPATH, value='/html/body/div[1]/div/div/div[6]/div[3]/div[3]/div[3]/a').click()
    time.sleep(2)
    source.append(html)
    print("Page -",i)
```

```python
dict={
    "mumbai":source
    
}
df=pd.DataFrame(dict)
df.to_csv('mumbai_phase-1.csv',index=False)
```
